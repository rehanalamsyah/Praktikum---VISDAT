**a. Scatter Plot dengan Matplotlib**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Membuat scatter plot
plt.scatter(suhu, penjualan)
plt.title("Hubungan Penjualan Es Krim dan Suhu")
plt.xlabel("Suhu (°C)")
plt.ylabel("Penjualan Es Krim")
plt.show()
```

**b. Kustomisasi Scatter Plot**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Kustomisasi scatter plot
plt.scatter(suhu, penjualan, color='orange', s=100, edgecolor='black', alpha=0.7)
plt.title('Hubungan Penjualan Es Krim dan Suhu')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.grid(True)
plt.show()
```

**c. Menambahkan Scatter Plot**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]

# Data tambahan untuk kategori hari
penjualan_kerja = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_akhir_pekan = [60, 70, 80, 100, 110, 120, 140, 160, 200]

# Scatter plot multiple
plt.scatter(suhu, penjualan_kerja, color='green', label='Hari Kerja', s=80)
plt.scatter(suhu, penjualan_akhir_pekan, color='purple', label='Akhir Pekan', s=80)
plt.title('Penjualan Es Krim Berdasarkan Hari')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.legend()
plt.show()
```

**d. Scatter Plot untuk Melihat Hubungan Data dengan Beberapa Variabel`**

```python
import matplotlib.pyplot as plt

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Data tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [200, 220, 240, 260, 280, 300, 320, 340, 360]

# Scatter plot dengan ukuran dan warna yang bervariasi
plt.scatter(suhu, penjualan, c=kelembapan, s=ukuran_titik, cmap='coolwarm', alpha=0.6)
plt.title('Hubungan Penjualan, Suhu, dan Kelembapan')
plt.xlabel('Suhu (°C)')
plt.ylabel('Penjualan Es Krim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data dummy
suhu = [20, 22, 24, 26, 28, 30, 32, 34, 36]
penjualan = [50, 60, 70, 90, 100, 110, 130, 150, 180]

# Data tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [200, 220, 240, 260, 280, 300, 320, 340, 360]

# Membuat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot dengan ukuran dan warna yang bervariasi
sc = ax.scatter(suhu, penjualan, kelembapan, c=kelembapan, s=ukuran_titik, cmap='coolwarm', alpha=0.6)

# Menambahkan label dan judul
ax.set_title('Hubungan Penjualan, Suhu, dan Kelembapan (3D)')
ax.set_xlabel('Suhu (°C)')
ax.set_ylabel('Penjualan Es Krim')
ax.set_zlabel('Kelembapan (%)')

# Menambahkan color bar
fig.colorbar(sc, label='Kelembapan (%)')

# Menampilkan plot
plt.show()
```
